```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
from matplotlib import animation
from IPython.display import HTML
```

```python
# Set the font family and size to use for Matplotlib figures.
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 12
```

```python
# Set the font family and size to use for Matplotlib figures.
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 12

# Set parameters.
L = 0.09  # total length of the rod
nx = 51  # total number of points
dx = L / (nx - 1)  # uniform spatial step

# **Material 1 properties**
rho1 = 1490  
k1 = 1.2  
Cp1 = 1050  
alpha1 = k1 / (rho1 * Cp1)  

# **Material 2 properties**
rho2 = 64 
k2 = 0.04 
Cp2 = 840  
alpha2 = k2 / (rho2 * Cp2)  

# **Determine interface location**
frac_mat1 = 2/3  # Fraction of the rod occupied by Material 1
interface_idx = int(frac_mat1 * nx)  # Index where material changes

# Define the locations along the rod.
x = numpy.linspace(0.0, L, num=nx)

# Set the initial temperature along the rod.
T0 = numpy.zeros(nx)
T0[:] = 293  # Initial uniform temperature
Tinf = 293
T0[0]=800
A = 0.1914  # Surface area
h = 30  # Convective heat transfer coefficient
boltzmann = 5.6704E-08  # Stefan-Boltzmann constant
Qdot_in = 1000  # Heat input
```

```python
def update_plot(n, T):
    """
    Update the line y-data of the Matplotlib figure.
    
    Parameters
    ----------
    n : integer
        The time-step index.
    T[n] : list of numpy.ndarray objects
        The history of the numerical solution.
    """
    fig.suptitle('Time step {:0>2} \n'.format(n),y=0.96)
    line.set_ydata(T[n])
```

```python
def ftcs_two_materials(T0, nt, dt, dx, alpha1, alpha2):
    """
    Computes and returns the temperature along the rod with two materials.
    
    Parameters
    ----------
    T0 : numpy.ndarray
        The initial temperature distribution.
    nt : int
        Number of time steps.
    dt : float
        Time step size.
    dx : float
        Grid spacing.
    alpha1 : float
        Thermal diffusivity of Material 1.
    alpha2 : float
        Thermal diffusivity of Material 2.
    
    Returns
    -------
    T_hist : list
        Temperature distribution history.
    """
    T = T0.copy()
    T_hist = [T0.copy()]
    
    for n in range(nt):
        T_new = T.copy()
        
        # **Update interior points for Material 1**
        for i in range(1, interface_idx):
            T_new[i] = T[i] + (alpha1 * dt / dx**2) * (T[i+1] - 2*T[i] + T[i-1])
        
        # **Update interior points for Material 2**
        for i in range(interface_idx + 1, nx - 1):
            T_new[i] = T[i] + (alpha2 * dt / dx**2) * (T[i+1] - 2*T[i] + T[i-1])
        
        # **Interface condition: Ensure heat flux continuity**
        k_left = k1
        k_right = k2
        T_new[interface_idx] = (k_left * T[interface_idx - 1] + k_right * T[interface_idx + 1]) / (k_left + k_right)

        # **Neumann boundary condition (heat loss at the right boundary)**
        Qdot_out = boltzmann * A * (T[-1]**4 - Tinf**4) + h * A * (T[-1] - Tinf)
        T_new[-1] = T[-1] + (alpha2 * dt / dx**2) * (T[-2] - T[-1] - heatfac * Qdot_out)
        
        # **Dirichlet boundary condition (heat input at the left boundary)**
        #T_new[0] = T[0] + (alpha1 * dt / dx**2) * (T[1] - T[0] + heatfac * Qdot_in)
        T_new[0] = T[0]
        
        T = T_new.copy()
        T_hist.append(T.copy())
    
    return T_hist
```

```python
# **Set the time-step size based on the CFL limit**
sigma = 0.5  # CFL safety factor
dt = sigma * dx**2 / max(alpha1, alpha2)  # Ensure stability

# **Limit dt to a maximum of 1s**
dt = min(dt, 1)
Tfinal = 6 * 3600  # Simulate for 6 hours
nt = int(Tfinal / dt)  # Number of time steps

# **Compute heat factor**
heatfac = dx / (k1 * A)

# **Run the simulation**
T = ftcs_two_materials(T0, nt, dt, dx, alpha1, alpha2)
```
