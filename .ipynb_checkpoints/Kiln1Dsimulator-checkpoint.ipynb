```python
import numpy as np
from matplotlib import pyplot
from matplotlib import animation
from IPython.display import HTML

# Set the font family and size to use for Matplotlib figures.
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 12
```

```python
# Compute thermal diffusivity for each segment based on temperature.
def compute_alpha(T):
    k = np.array([k_funcs[i](T[i]) for i in range(nx)])
    Cp = np.array([Cp_funcs[i](T[i]) for i in range(nx)])
    return k / (rho * Cp), k, Cp
```

```python
# FTCS solver with mixed boundary conditions
def ftcs_mixed_bcs(T0, nt, dt, dx):
    """
    Computes and returns the temperature along the rod
    after a provided number of time steps,
    given the initial temperature and the thermal diffusivity.
    The diffusion equation is integrated using forward 
    differencing in time and central differencing in space.
    The function uses a Dirichlet condition on the left side
    of the rod and a Neumann condition (zero-gradient) on the
    right side.
    
    Parameters
    ----------
    T0 : np.ndarray
        The initial temperature along the rod as a 1D array of floats.
    nt : integer
        The number of time steps to compute.
    dt : float
        The time-step size to integrate.
    dx : float
        The distance between two consecutive locations.
    
    Returns
    -------
    T : np.ndarray
        The temperature history as a list of 1D arrays of floats.
    """
    T = T0.copy()
    T_hist = [T0.copy()]
    for n in range(nt):
        alpha, k, Cp = compute_alpha(T)
        T_new = T.copy()
        for i in range(1, nx - 1):
            sigma = alpha[i] * dt / dx**2
            T_new[i] = T[i] + sigma * (T[i + 1] - 2.0 * T[i] + T[i - 1])

        # Apply boundary conditions
        Qdot_out = boltzmann * A * (T[-1]**4 - Tinf**4) + h * A * (T[-1] - Tinf)
        heatfac = dx / (k[-1] * A)
        T_new[-1] = T[-1] + alpha[-1] * dt / dx**2 * (T[-2] - T[-1] - heatfac * Qdot_out)
        
        heatfac_in = dx / (k[0] * A)
        T_new[0] = T[0] + alpha[0] * dt / dx**2 * (T[1] - T[0] + heatfac_in * Qdot_in)

        T_hist.append(T_new.copy())
        T = T_new.copy()

    return T_hist
```

```python
# Set parameters.
L = 0.1  # length of the rod
nx = 101  # number of locations on the rod
dx = L / (nx - 1)  # distance between two consecutive locations

# Define spatially varying material properties.
# Materials: [Material1, Material2, ...]


materials = [
    {'rho': 1490, 'k': lambda T: 1.2, 'Cp': lambda T: 1050},
    {'rho': 64, 'k': lambda T: 0.04, 'Cp': lambda T: 840},
]

"""
materials = [
    {'rho': 1490, 'k': lambda T: 1.2, 'Cp': lambda T: 1050},
    {'rho': 1490, 'k': lambda T: 1.2, 'Cp': lambda T: 1050},
]
"""
"""
materials = [
    {'rho': 1490, 'k': lambda T: 1.148 + 0.0005315 * (T - 293), 'Cp': lambda T: 753.622+1.55*(T-293)-0.001403*(T-293)**2},
    {'rho': 1490, 'k': lambda T: 0.00000025*(T-293)**2+0.00011*(T-293)+0.044, 'Cp': lambda T: 840},
]
"""

"""
materials = [
    {'rho': 1490, 'k': lambda T: 1.148 + 0.0005315 * (T - 293), 'Cp': lambda T: 753.622+1.55*(T-293)-0.001403*(T-293)**2},
    {'rho': 200, 'k': lambda T: 0.031675-6.80767e-5*(T-293)+3.007e-7*(T-293)**2, 'Cp': lambda T: 840},
]
"""

# Assign material properties based on regions.
region_boundary = [0.06]  # Boundary at x=0.03 (halfway point)

rho = np.zeros(nx)
k_funcs = []
Cp_funcs = []

x = np.linspace(0.0, L, num=nx)
for i in range(nx):
    if x[i] < region_boundary[0]:
        rho[i] = materials[0]['rho']
        k_funcs.append(materials[0]['k'])
        Cp_funcs.append(materials[0]['Cp'])
    else:
        rho[i] = materials[1]['rho']
        k_funcs.append(materials[1]['k'])
        Cp_funcs.append(materials[1]['Cp'])


# Set the initial temperature along the rod.
T0 = np.zeros(nx)
T0[:] = 293
Tinf = 293

A = 0.1914  # Cross-sectional area of the rod
h = 30  # Convective heat transfer coefficient
boltzmann = 5.6704E-08  # Stefan-Boltzmann constant
Qdot_in = 2500  # Heat input at the left boundary




# Set the time-step size based on CFL limit.
sigma = 0.5
dt_stable = sigma * dx**2 / max(k_funcs[0](293) / (rho * Cp_funcs[0](293)))  # time-step size

# Compute the temperature along the rod.
dt = min(dt_stable, 1.0)  # Ensure the time step is within stable limits
Tfinal = 6 * 3600  # Simulation duration (seconds)
nt = int(Tfinal / dt)

T = ftcs_mixed_bcs(T0, nt, dt, dx)

# Plot the initial temperature profile
fig = pyplot.figure(figsize=(6.0, 4.0))
pyplot.xlabel('Distance [m]')
pyplot.ylabel('Temperature [K]')
pyplot.grid()
line = pyplot.plot(x, T[-1],
                   color='C0', linestyle='-', linewidth=2)[0]
pyplot.ylim(0.0, max(T[-1]) + 50)
pyplot.tight_layout()



print(dt)
```

```python
# Extract the first number of each subarray
first_elements = [subarray[0] for subarray in T]

# Print extracted elements for verification
#print("First elements of each subarray:", first_elements)
time = (np.linspace(0,nt*dt,nt+1))/3600
# Plotting the values
pyplot.plot(time,first_elements, linestyle='-')

pyplot.xlabel('time[h]')
pyplot.ylabel('Tint [K]')
pyplot.title('Temperature inside the Kiln vs Time')
pyplot.grid(True)
pyplot.show()
```

```python
last_elements = [subarray[-1] for subarray in T]

pyplot.plot(time,last_elements, linestyle='-',c='r')
pyplot.xlabel('time[h]')
pyplot.ylabel('Tint [K]')
pyplot.title('Temperature exterior of the Kiln vs Time')
pyplot.grid(True)
pyplot.show()
```

```python
# Function to update the plot
def update_plot(n, T):
    """
    Update the line y-data of the Matplotlib figure.
    
    Parameters
    ----------
    n : integer
        The time-step index.
    T : list of np.ndarray objects
        The history of the numerical solution.
    """
    fig.suptitle('Time {0:.2f} [min]'.format(n*200/60))
    line.set_ydata(T[n])
    #line.set_xdata(time)
```
